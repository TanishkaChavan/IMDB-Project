```python
q = ty.groupby( ['original_language']).agg( {'revenue' : 'sum' , 'budget': 'sum'})
```

```python
q.reset_index( inplace = True)
```

```python
q = q[q['original_language'] != 'en']
```

```python
# Create a stacked bar plot
plt.figure(figsize=(8, 5))
plt.bar(q['original_language'],q['revenue'] , color='lightblue', edgecolor='black', label='revenue')
plt.bar(q['original_language'], q['budget'], bottom=q['revenue'], color='salmon', edgecolor='black', label='budget')
plt.xticks( rotation = 90)
```
